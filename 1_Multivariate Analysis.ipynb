Grouped boxplot are used when you have a numerical variable, several groups and subgroups.

It is easy to realize one using seaborn.Y is your numerical variable, x is the group column, and hue is the subgroup column.

```python
# library and data
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
df = sns.load_dataset('iris')
 
# # plot
f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)
# print (axes)
sns.distplot( df["sepal_length"] , color="skyblue", ax=axes[0, 0])
# sns.distplot( df["sepal_width"] , color="olive", ax=axes[0, 1])
sns.boxplot( df["sepal_width"] , color="olive", ax=axes[0, 1])
sns.distplot( df["petal_length"] , color="gold", ax=axes[1, 0])
sns.distplot( df["petal_width"] , color="teal", ax=axes[1, 1])
```

## Heat Map

A heat map (or heatmap) is a graphical representation of data where the individual values contained in a matrix are represented as colors. It is a bit like looking a data table from above. It is really useful to display a general view of numerical data,

not to extract specific data point. It is quite straight forward to make a heat map,

- matrix where each row is an individual, and 
- each column represents an observation.

a heatmap consists to make a visual representation of the matrix: 
- each square of the heatmap represents a cell. 
- The color of the cell changes following its value.



- Suppose you measured several variables for n individuals. 
- A common task is to check if some variables are correlated. 
- You can easily calculate the correlation between each pair of variable, and plot this as a heatmap. 
- This lets you discover which variable are related one each other.

```python
%matplotlib inline
from ggplot import mtcars 
import seaborn as sns
# print (mtcars.head())
print (mtcars.corr())
a = mtcars.corr()
sns.heatmap(a)
```

```python
sns.heatmap(a)
```

### Annotate each cell with value
### annot=True will add a label into each cell, providing the exact value behind the color.

## Basic Bubble plot

### You need 3 numerical variables as input: one is represented by the X axis, one by the Y axis, and one by the size.

- A bubble plot is very close to a scatterplot. 
- With Matplotlib, we will construct them using the same scatter function. 
- However, we will use the s argument to map a third numerical variable to the size of the marker. This will give us an additive information on the same graphic!
- A bubble chart is a type of chart that displays three dimensions of data. 
- Each entity with its triplet (v1, v2, v3) of associated data is plotted as a disk that expresses two of the vi values through the disk's xy location and the third through its size. 

- Note that too many bubble make the chart hard to read, so this type of representation is usually not recommended for big amount of data. 

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('titanic')
print (df.head(2))

plt.scatter(df['age'],df['pclass'])
plt.scatter(df['age'],df['pclass'], s = df['fare']) # s is size it is relative size
```

### Analysis:
- third parameter is showing the fare details and it shows the fare in the 1st class is relatively  high when compared to the prices in the class 2 and class 3.

## The Car Accidents Injuries Rate in 5 Canada Provinces

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm # colormap
import numpy as np

province=['Ontario','Quebec','BritishColumbia','Manitoba','NovaScoti']
size = [908.607,1356.547,922.509,552.329,651.036]  # area of the province  Size(*1000km2)
population = [12851821,7903001,4400057,1208268,4160000] #Population(ten million)
injuries = [625,752,629,1255,630]

# # # Choose some random colors

colors=cm.rainbow(np.random.rand(5))#generate 5 random colors
# # Use those colors as the color argument
#plt.scatter(size,population)
plt.scatter(size,population,s=injuries,color=colors)
N=5

for i in range(N):
#     print(province[i])
#     xy=(size[i],population[i])
#     print(xy)
    plt.annotate(province[i],xy=(size[i],population[i]))
plt.xlabel('Size(*1000km2)')
plt.ylabel('Population(ten million)')

# # # Move title up with the "y" option
plt.title('The Car Accidents Injuries Rate in 5 Canada Provinces',y=1.05)
# # plt.show()
```
